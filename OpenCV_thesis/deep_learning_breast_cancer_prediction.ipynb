```python
from keras.models import Sequential
from keras.layers import Dense, Dropout
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import scale, LabelEncoder

data = pd.read_csv('bc_data.csv')

X = data.iloc[:,2:-1]
y = data.iloc[:,1]

enc = LabelEncoder()
y = enc.fit_transform(y)
```

```python
X = scale(X)
##
#print(X)
##
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
##
model = Sequential()
model.add(Dense(12, kernel_initializer = 'uniform', input_dim = 30, activation = 'relu'))
model.add(Dropout(.5))
model.add(Dense(10, kernel_initializer = 'uniform', activation = 'relu'))
model.add(Dropout(.5))
model.add(Dense(6, kernel_initializer = 'uniform', activation = 'relu'))
model.add(Dropout(.5))
model.add(Dense(1, kernel_initializer = 'uniform', activation = 'sigmoid'))
```

```python
model.compile(loss = 'mean_squared_error', optimizer = 'adam', metrics = ['accuracy'])
model.fit(X_train, y_train, epochs = 50, batch_size = 10, verbose = 2)
##
preds = model.predict(X_test)
##
bin_preds = [round(x[0]) for x in preds]
##
###print(bin_preds)
count = 0
for i in range(len(bin_preds)):
    if bin_preds[i] == y_test[i]:
        count += 1
##
accur = count/(len(y_test))
##
print(accur)
```
