```python
import pandas as pd
from operator import itemgetter
import scipy as sp
import numpy as np
from sklearn import preprocessing
from scipy.stats.stats import pearsonr  
orig_data = pd.read_csv('/Users/nthor/Desktop/hackathon_final.csv')
print(orig_data)
```

```python
comp = orig_data.iloc[0,5]
new_data = orig_data[orig_data['Current_Company'] == comp]
print(new_data)
```

```python
data = new_data.iloc[:,5:-2]
print(data)
```

```python
data = data.apply(lambda x: (x - np.mean(x)) / (np.max(x) - np.min(x)))
data.dropna(axis = 1, inplace = True)
print(data)
```

```python
corrs = []
for i in range(1,(len(data))):
    corrs.append((pearsonr(data.iloc[i,:], data.iloc[0,:])))

print(corrs)    
val_1 = max(corrs,key=itemgetter(1))
val_2 = max(corrs,key=itemgetter(0))
#indices = []
#indices.append([i for i, tupl in enumerate(corrs) if tupl == val_1])
indices = [i for i, tupl in enumerate(corrs) if tupl == val_2]
indices_ = [i for i, tupl in enumerate(corrs) if tupl == val_1]
print(new_data.iloc[indices[0]+1, :])
print(new_data.iloc[indices_[0]+1, :])
print(new_data.iloc[4, :])

```
